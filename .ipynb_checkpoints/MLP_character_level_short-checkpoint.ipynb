```python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Read in all the words
words = open('names.txt', 'r').read().splitlines()
words[:8]
```

```python
len(words)
```

```python
# Build the vocabulary of characters and mapping to/from integers
chars = sorted(list(set(''.join(words)))) # All individual words in the dataset
stoi = {s:i+1 for i,s in enumerate(chars)} # String to integer mapping
stoi['.'] = 0 # In the vocabulary doen't exist the '.' character by default, we include it at the first position
itos = {i:s for s,i in stoi.items()} # Integer to string mapping
print(itos)
```

```python
characters_integer = []
for ch in 'emma':
    characters_integer.append(stoi[ch])

print(torch.tensor(characters_integer))
```

```python
# Build the dataset

block_size = 3 # Context length: how many characters do we take to predict the next one
X, Y = [], []
for w in words:
    #print(w)
    # Creates a list with three elements, each of which is 0
    # This context list will be a sliding window always containing the previous three characters
    # in order to predict the next
    context = [0] * block_size
    for ch in w + '.':
        ix = stoi[ch]
        # Our input data now will be a context of size three different to the context fo size one we builded
        # in the previous code lab
        X.append(context)
        Y.append(ix)
        #print(''.join(itos[i] for i in context), '--->', itos[ix])
        context = context[1:] + [ix]
        
X = torch.tensor(X)
Y = torch.tensor(Y)
```

```python
# Dataset
X.shape, Y.shape 
```

```python
g = torch.Generator().manual_seed(2147483647)
C = torch.randn((27, 2), generator=g)
W1 = torch.randn((6, 100), generator=g)
b1 = torch.randn(100, generator=g)
W2 = torch.randn((100, 27), generator=g)
b2 = torch.randn(27, generator=g)
parameters = [C, W1, b1, W2, b2]
```

```python
# Number of parameters in total
sum(p.nelement() for p in parameters)
```

### Forward pass

```python
# F.cross_entropy(logits, Y) is equivalent to
# counts = logits.exp()
# prob = counts / counts.sum(1, keepdims=True)
# loss = -prob[torch.arange(32), Y].log().mean()
```

```python
emb = C[X] # (32, 3, 2)
h = torch.tanh(emb.view(emb.shape[0], 6) @ W1 + b1) # (32, 100)
logits = h @ W2 + b2 # (32, 27)
loss = F.cross_entropy(logits, Y)
loss
```

```python
# There are many good reasons to call cross-entropy.
# 1. Forward pass can be much more efficient, and also the backward pass can be much more efficient
# 2. Things can be much more numerically well behaved
```

### Forward, Backward propagation and parameters updated

```python
for param in parameters:
    param.requires_grad = True
```

```python
for _ in range(100):
    # Minibatch construct
    ix = torch.randint(0, emb.shape[0], (32, ))
    
    # Forward pass
    emb = C[X[ix]] # (32, 3, 2)
    h = torch.tanh(emb.view(emb.shape[0], 6) @ W1 + b1) # (32, 100)
    logits = h @ W2 + b2 # (32, 27)
    loss = F.cross_entropy(logits, Y[ix])
    
    # Backward pass
    # We want to set the gradients to be zero
    for param in parameters:
        param.grad = None

    loss.backward()
    
    # Update the parameters
    for param in parameters:
        param.data += -0.1 * param.grad
        
print(loss.item())
```

```python
# Finds the maximum values along along with their indices in each row
# (dimension 1, chich corresponds to columns)
logits.max(1)
```
