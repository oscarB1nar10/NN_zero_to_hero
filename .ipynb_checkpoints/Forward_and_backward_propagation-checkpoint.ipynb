### Derivative overview

```python
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Function to derive 3x^2 - 4x + 5
# Its derivative is 6x - 4
def f(x):
    return 3*x**2 - 4*x + 5
```

```python
# Plotting the function
xs = np.arange(-5, 5, 0.25)
ys = f(xs)
plt.plot(xs, ys)
```

```python
# Where f(a) is the ouput of evaluating the input a in f(a)
# f(a + h) is the output of evaluation the input a + delta-x in f(a). delta-x is a small value which tends to zero
h = 0.001
a = 2
derivative_numerical_estimation = (f(a + h) - f(a)) / h

print(f"Exact derivative value at a=2: 6*(a) - 4 => 8")
print(f"Numerical estimation of the derivative: {derivative_numerical_estimation}")
```

```python
# Let's try with a more complex function of 3 variables
a = 2.0
b = -3.0
c = 10.0
d = a*b + c

print(d)
```

```python
# Let's apply some differential to the variable a
# Notice how the value of slope is basically the value of b (The partial derivative of d2 with respect to a)
h = 0.01

a = 2.0
b = -3.0
c = 10.0

d1 = a*b + c
a += h 
d2 = a*b + c

print(f'd1: {d1}')
print(f'd2: {d2}')
print(f'slope: {(d2 - d1) / h}')
```

```python
# Notice how the value of slope is basically the value of a (The partial derivative of d2 with respect to b)
a = 2.0
b = -3.0
c = 10.0

d1 = a*b + c
b += h 
d2 = a*b + c

print(f'd1: {d1}')
print(f'd2: {d2}')
print(f'slope: {(d2 - d1) / h}')
```

```python
# Notice how the value of slope is 1 (The partial derivative of d2 with respect to c)
a = 2.0
b = -3.0
c = 10.0

d1 = a*b + c
c += h 
d2 = a*b + c

print(f'd1: {d1}')
print(f'd2: {d2}')
print(f'slope: {(d2 - d1) / h}')
```

### Math graph, Forward and Backward propagation

```python
# Class container for operations in the math graph
# _children will be a tuple
class Value:
    
    def __init__(self, data, _children=(), _op='', label=''):
        self.data = data
        self.grad = 0.0
        # It is going to store the chain rule at each node
        self._backward = lambda: None
        self._prev = set(_children)
        self._op = _op
        self.label = label
        
    def __repr__(self):
        return f"Value(data={self.data})"
    
    def __add__(self, other):
        out = Value(self.data + other.data, _children=(self, other), _op='+')
        
        # Here we are going to implement the derivative of the 'parent' node (neuron)
        # with respect to the chield nodes
        # p = parent, c1 = chield1, c2 = chield2
        # p = c1 + c2
        # dp/dc1 = 1 = dp/dc2
        # L = lost function (output of the last neuron)
        # dL/dc1 = dL/dp * dp/dc1
        def _backward():
            self.grad = 1.0 * out.grad
            other.grad = 1.0 * out.grad
            
        out._backward = _backward
        return out
    
    def __mul__(self, other):
        out = Value(self.data * other.data, _children=(self, other), _op='*')
        return out
    
    def tanh(self):
        x = self.data
        t = (math.exp(2*x) - 1)/(math.exp(2*x) + 1)
        out = Value(t, _children=(self, ), _op='tanh')
        return out
    
a = Value(data=2.0, label='a')
b = Value(data=-3.0, label='b')
c = Value(10.0, label='c')
e = a*b; e.label = 'e'
d = e+c; d.label = 'd'
f = Value(-2.0, label='f')
L = d*f; L.label='L'
L
```

```python
from graphviz import Digraph

def trace(root):
    # builds a set of all nodes and edges in a graph
    nodes, edges = set(), set()
    
    def build(node):
        # Add the current node to the nodes if not present
        if node not in nodes:
            nodes.add(node)
            print(f"{node.data}, child: {node._prev}, operation: {node._op}")
        # For every child of node save the child and node in edges, and add
        # the child as other node calling build(child)
        for child in node._prev:
            edges.add((child, node))
            build(child)
            
    build(root)
    return nodes, edges


def draw_dot(root):
    dot = Digraph(format='svg', graph_attr={'rankdir': 'LR'}) # LR = left to right
    
    nodes, edges = trace(root)
    print("\n")
    print(f"nodes: {nodes}\nedges: {edges}")
    for nod in nodes:
        uid = str(id(nod))
        # For any value in the graph, create a rectangular 'record' node for it
        dot.node(name=uid, label="{ %s | data %.4f | grad %.4f }" % (nod.label, nod.data, nod.grad), shape='record')
        if nod._op:
            # If this value is a result of some operation, create an op node for it
            dot.node(name=uid + nod._op, label=nod._op)
            # and connect this node to it
            dot.edge(uid + nod._op, uid)
            
    for nod1, nod2 in edges:
        # Connect nod1 to the op node of nod2
        dot.edge(str(id(nod1)), str(id(nod2)) + nod2._op)
    
    return dot
        
    
```

```python
# Forward pass
draw_dot(L)
```

```python
# In the backward propagation we are going to calculate the derivate of each node with respect to L
# So we are going to calculate the derivative of L with respect to {d, f, e, c, b, a}
# Here L mimics the lost function of a neural network while the other nodes values {d, f, e, c, b, a} mimic
# the weights of the neural network
```

```python
# Grad of L with respect to L
def grad_calcutator():
    h = 0.001
    
    a = Value(data=2.0, label='a')
    b = Value(data=-3.0, label='b')
    c = Value(10.0, label='c')
    e = a*b; e.label = 'e'
    d = e+c; d.label = 'd'
    f = Value(-2.0, label='f')
    L = d*f; L.label='L'
    L1 = L.data
    
    a = Value(data=2.0, label='a')
    b = Value(data=-3.0, label='b')
    c = Value(10.0, label='c')
    e = a*b; e.label = 'e'
    d = e+c; d.label = 'd'
    f = Value(-2.0, label='f')
    L = d*f; L.label='L'
    L2 = L.data + h
    
    print((L2 - L1)/h)
    
grad_calcutator()
    
```

```python
# Grad of L with respect to d
# L = d*f
# dl / dd = f = -2.0
# And by simmetry dl / df = d = 4
def grad_calcutator():
    h = 0.001
    
    a = Value(data=2.0, label='a')
    b = Value(data=-3.0, label='b')
    c = Value(10.0, label='c')
    e = a*b; e.label = 'e'
    d = e+c; d.label = 'd'
    f = Value(-2.0, label='f')
    L = d*f; L.label='L'
    L1 = L.data
    
    a = Value(data=2.0, label='a')
    b = Value(data=-3.0, label='b')
    c = Value(10.0, label='c')
    e = a*b; e.label = 'e'
    d = e+c; d.label = 'd'
    d.data += h
    f = Value(-2.0, label='f')
    L = d*f; L.label='L'
    L2 = L.data
    
    print((L2 - L1)/h)
    
grad_calcutator()
```

```python
d.grad = -2.0
f.grad = 4
```

```python
# Grad of L with respect to c
# (dl/dc) = (dl/dd) * (dd/dc)
# (dl/dc) = -2.0 * (dd/dc)
# d = e+c => (dd/dc) = 1
# (dl/dc) = -2
# And by symetry (dd/de) = -2
def grad_calcutator():
    h = 0.001
    
    a = Value(data=2.0, label='a')
    b = Value(data=-3.0, label='b')
    c = Value(10.0, label='c')
    e = a*b; e.label = 'e'
    d = e+c; d.label = 'd'
    f = Value(-2.0, label='f')
    L = d*f; L.label='L'
    L1 = L.data
    
    a = Value(data=2.0, label='a')
    b = Value(data=-3.0, label='b')
    c = Value(10.0, label='c')
    c.data +=h
    e = a*b; e.label = 'e'
    d = e+c; d.label = 'd'
    f = Value(-2.0, label='f')
    L = d*f; L.label='L'
    L2 = L.data
    
    print((L2 - L1)/h)
    
grad_calcutator()
```

```python
c.grad = -2
e.grad = -2
```

```python
# Grad of L with respect to a
# (dl/da) = (dd/de) * (de/da)
# (dl/da) = -2.0 * (de/da)
# e = a*b => (de/da) = b = -3
# (dl/da) = -2.0 * -3 = 6
def grad_calcutator():
    h = 0.001
    
    a = Value(data=2.0, label='a')
    b = Value(data=-3.0, label='b')
    c = Value(10.0, label='c')
    e = a*b; e.label = 'e'
    d = e+c; d.label = 'd'
    f = Value(-2.0, label='f')
    L = d*f; L.label='L'
    L1 = L.data
    
    a = Value(data=2.0, label='a')
    a.data+=h
    b = Value(data=-3.0, label='b')
    c = Value(10.0, label='c')
    e = a*b; e.label = 'e'
    d = e+c; d.label = 'd'
    f = Value(-2.0, label='f')
    L = d*f; L.label='L'
    L2 = L.data
    
    print((L2 - L1)/h)
    
grad_calcutator()
```

### Backpropagation in a Simple NN

```python
# tanh activation function
plt.plot(np.arange(-5, 5, 0.2), np.tanh(np.arange(-5, 5, 0.2)))
plt.grid()
```

```python
# Inputs x1,x2
x1 = Value(2.0, label='x1')
x2 = Value(0.0, label='x2')
# weights w1,w2
w1 = Value(-3.0, label='w1')
w2 = Value(1.0, label='w2')
# bias of the neuron
b = Value(6.8813735870195432, label='b')

# Dot product between weights, inputs and bias
# output = (weights . inputs) + bias
# w1x1 + w2x2 + b 
x1w1 = x1*w1; x1w1.label='x1w1'
x2w2 = x2*w2; x2w2.label='x2w2'
x1w1x2w2 = x1w1 + x2w2; x1w1x2w2.label='x1*w1 + x2*w2'
n = x1w1x2w2 + b; n.label='n'
o = n.tanh(); o.label='o'
```

```python
draw_dot(o)
```

```python
# Let's start backpropagation for the neuron
# The bellow formula is the derivative of the operation tanh(x)
# I'ts important to know what is the derivative of the operations in order to do the backpropagation
```

```python
# do/do = 1.0
o.grad = 1.0
```

```python
# o = tanh(n)
# do/dn = 1 - tanh(n)^2
# do/dn = 1 - o**2
# d0/dn = 1 - 0.5 = 0.5
n.grad = 0.5
```

```python
#NOTE: A plus operation is a distributor of gradient

# Let r = (x1*w1 + x2*w2)
# n = r + b
# dn/dr = 1
# dn/db = 1
# do/dr = do/dn * dn/dr
# do/dr = 0.5 * 1 = 0.5
x1w1x2w2.grad = 0.5
b.grad = 0.5
```

```python
# For both the gradient will be 0.5 due that the operation is plus and it is a distributor of gradient
# s = x1w1 = x1*w1
# p = x2w2 = x2*w2
# r = s + p
# dr/ds = 1
# dr/dp = 1
# do/ds = dn/dr * dr/ds = 0.5
# do/dp = dn/dr * dr/dp = 0.5
x1w1.grad = 0.5
x2w2.grad = 0.5
```

```python
# s = x1w1 = x1*w1
# ds/w1 = x1
# ds/x1 = w1
# do/w1 = dr/ds * ds/w1
# do/w1 = 0.5 * 2.0 = 1
# do/X1 = dr/ds * ds/x1
# do/w1 = 0.5 * -3.0 = -1.5
w1.grad = 1.0
x1.grad = -1.5
```

```python
# p = x2w2 = x2*w2
# dp/w2 = x2
# dp/x2 = w2
# do/w2 = dr/dp * dp/w2
# do/w2 = 0.5 * 0 = 0
# do/X2 = dr/dp * dp/x2
# do/X2 = 0.5 * 1.0 = 0.5
w2.grad = 0
x2.grad = 0.5
```
